```python
#!pip install sodapy
import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cdc.gov", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cdc.gov,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("8pt5-q6wp", limit=5000)

# Convert to pandas DataFrame
df = pd.DataFrame.from_records(results)
```

```python
df
```

```python
age = df[df['group']=='By Age']
ny = df[df['state']=='New York']
ny1 = df[df['state']=='New York']
ny2 = df[df['state']=='New York']
ny3 = df[df['state']=='New York']
ny4 = df[df['state']=='New York']
ny5 = df[df['state']=='New York']
ny6 = df[df['state']=='New York']
ny.reset_index(drop=True, inplace=True)
ny1.reset_index(drop=True, inplace=True)
ny2.reset_index(drop=True, inplace=True)
ny3.reset_index(drop=True, inplace=True)
ny4.reset_index(drop=True, inplace=True)
ny5.reset_index(drop=True, inplace=True)
ny6.reset_index(drop=True, inplace=True)

#combined = pd.concat([age,ny])
ageRange = ['18 - 29 years', '30 - 39 years', '40 - 49 years', '50 - 59 years', '60 - 69 years', '70 - 79 years', '80 years and above']
ny['age'] = ageRange[0]
ny1['age'] = ageRange[1]
ny2['age'] = ageRange[2]
ny3['age'] = ageRange[3]
ny4['age'] = ageRange[4]
ny5['age'] = ageRange[5]
ny6['age'] = ageRange[6]


ny6
```

```python
new_york = pd.concat([ ny, ny1, ny2, ny3, ny4, ny5, ny6])
new_york.to_csv('NY_MENTAL_HEALTH_ONLY.csv', index=False)
```

```python
def standard (df):
    helper = df
    totals = helper.loc['0-4'] + helper.loc['5-12'] + helper.loc['13-17']
    totals1 = helper.loc['18-24'] + helper.loc['25-34'] + helper.loc['35-44']
    totals2 = helper.loc['45-54'] + helper.loc['55-64']
    totals3 = helper.loc['65-74']
    totals4 = helper.loc['75+']
    citywide = helper.loc['Citywide']
    
    df = pd.concat([totals, totals1, totals2, totals3, totals4, citywide], axis=1)
    df = df.T
    df = df.rename_axis("AGE_GROUP")
    df_new = df.rename(index={0: '0-17', 1: '18-44', 2: '45-64'})
    return df_new
```
